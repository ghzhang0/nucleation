```python
# FIRE parameters (suggested by bitzek et. al)
alpha_s = 0.1
nmin = 5
finc = 1.1
fdec = 0.5
falph = 0.99
# output energy every dumpstep steps 
dumpstep = 100
# Tunable FIRE params: dt should be 10 dt_MD
dtmax =  1e-3
# tolerance for max gradient element
crit  = 1e-2
```

```python
# Lattice parameters
n_init = 12
ctheta = np.pi/3
n_rings = 2

# Bladon-Frenkel potential parameters
epsilon = 1.0
delsig = 0.5
```

```python
latt = Crystal(n_init, n_rings, ctheta)
potential = BF(epsilon, delsig)
fire = FIRE(alpha_s, falph, dtmax, finc, fdec, dumpstep, crit, nmin)
```

```python
energies, positions = fire.run(latt, potential)
```
